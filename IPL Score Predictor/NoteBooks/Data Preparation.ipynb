```python
for i in l:
    rows=data[data["striker"]==i].sort_values(by="match_id",ascending=False)[:10]
    runs=[]
    for k in rows["runs"]:
        runs.append(k)
    avg=sum(runs)/len(runs)
    for j in range(len(runs),10):
        runs.append(0)
    b=[]
    for k in rows["balls"]:
        b.append(k)
    for j in range(len(b),10):
        b.append(0)
    row=[i]
    
    sr=[]
    for j in range(10):
        if b[j]!=0:
            sr.append(runs[j]/b[j])
        else:
            sr.append(0)
    
    row.extend(runs)
    row.extend([avg])
    row.extend(b)
    row.extend(sr)
    form.loc[len(form)]=row
```

```python
team_players=[]
for item in teams:
    curr = dff[dff['batting_team']==item]
    last_match = sorted(curr['match_id'].unique(), reverse=True)[0]
    lastm = dff[ dff['match_id']==last_match ]
    
    players = [item]
    
    batters = lastm[lastm['batting_team']==item]
    batters = batters['striker'].unique()
    players.extend(batters)
    
    bowlers = lastm[lastm['bowling_team']==item]
    bowlers = bowlers['bowler'].unique()
    
    mask = np.isin(bowlers, batters)
    dis_bowlers = bowlers[~mask]
    players.extend(dis_bowlers)

    team_players.append(players)

team_players
```
